```python
import seaborn as sns
```

```python
health = sns.load_dataset("healthexp")
health.head()
```

### Exercise 7: Missing Data
1. Identify and count the missing values in the DataFrame from Exercise 3.
2. Remove the rows with missing values or fill the missing values with appropriate values.

```python
autos.isna().sum()
```

```python
autos[autos['bore'].isnull()]
```

```python
autos.query("`make` == 'mazda'")
```

```python
mean_bore_mazdas = autos[autos['make'] == 'mazda']['bore'].median()
mean_bore_mazdas
```

```python
autos[autos['bore'].isnull()] = mean_bore_mazdas
autos.loc[autos["make"] == 'mazda', 'bore'] = autos.loc[autos["make"] == 'mazda', 'bore'].fillna(mean_bore_mazdas)
```
